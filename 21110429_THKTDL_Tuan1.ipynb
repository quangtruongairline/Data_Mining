#### 10 equi-depth ranges

```python
import numpy as np

def equi_depth(data, k):

    sorted_data = np.sort(data)
    
    # Tinh so luong phan tu trong moi bins
    bin_size = len(data) // k
    
    # so luong phan tu thua de chia vao cac khoang bin
    remainder = len(data) % k
    
    # mang luu label cua bin
    labels = np.zeros(len(data))
    
    label = 0
    i = 0
    while i < len(data):
        labels[i:i+bin_size] = label
        i += bin_size
        if remainder > 0:
            i += 1
            remainder -= 1
        label += 1
    
    return labels.astype(int)
```

```python
# feature tuoi
equi_depth(df[0],10)
```

```python
# vẽ dữ liệu feature "tuổi" sau khi rời rạc hóa equi-depth
import matplotlib.pyplot as plt
plt.hist(equi_depth(df[0],10),bins=10)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

#### PCA

```python
from sklearn.decomposition import PCA
```

##### Chọn số chiều n_components để thuật toán giữ lại thông tin tốt

```python
# PCA với số chiều chưa xác định
```

```python
pca=PCA(n_components=None)
data_PCA=pca.fit_transform(data_scaler)
```

```python
pca.n_components_
```

```python
#eigenvector
pca.components_
```

Cụ thể, ma trận pca.components_ có kích thước là (n_components, n_features), với n_components là số lượng thành phần chính đã được giảm chiều, n_features là số lượng biến ban đầu. 

Mỗi vector cột của pca.components_ là một thành phần chính, được sắp xếp theo độ quan trọng giảm dần. Cụ thể, thành phần chính đầu tiên (vị trí đầu tiên trong pca.components_) sẽ giải thích phương sai lớn nhất của dữ liệu, và tiếp theo là các thành phần chính khác, giải thích phương sai giảm dần.

```python
#eigenvalues
pca.explained_variance_
```

Các giá trị eigenvalue trong pca.explained_variance_ được sắp xếp theo thứ tự giảm dần. Giá trị eigenvalue lớn hơn cho biết mức độ quan trọng của principal component tương ứng. Giá trị eigenvalue lớn nhất tương ứng với principal component đầu tiên và được sử dụng để giảm số chiều của dữ liệu.


```python
variance_ratio = pca.explained_variance_ratio_

cumulative_variance_ratio = np.cumsum(variance_ratio)
```

Trong PCA, explained_variance_ratio_ là một mảng các tỉ lệ phương sai được giải thích bởi từng thành phần chính. Mỗi phần tử trong mảng này thể hiện đóng góp của mỗi thành phần chính vào việc giải thích phương sai của dữ liệu.

```python
variance_ratio #ty le phuong sai
```

```python
cumulative_variance_ratio #Tong tich luy
```

### Trực quan với số chiều chưa xác định

```python
import matplotlib.pyplot as plt

plt.plot(range(1, len(cumulative_variance_ratio) + 1), cumulative_variance_ratio)
plt.title('Cumulative Variance Ratio')
plt.xlabel('Number of Principal Components')
plt.ylabel('Cumulative Variance Ratio')
plt.show() 
```

```python
import matplotlib.pyplot as plt
plt.bar(range(1, pca.n_components_ + 1), pca.explained_variance_ratio_)
plt.xlabel('Number of Components')
plt.ylabel('Explained Variance Ratio')
plt.title('Scree Plot')
plt.show()
```
