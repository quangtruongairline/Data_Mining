## Tính các cột còn lại


```python
# Hàm tính chuẩn k , trả về ma trận, với a[i,j] của ma trận là khoảng cách chuẩn k của feature i và j
# nếu i=j thì bằng 0 
def distp(k):
    x,y=df.shape
    dist_p=np.zeros((y,y))
    for i in range(0,y):
        for j in  range(0,y):
            dist_p[i,j]=np.linalg.norm(df[i]-df[j],ord=k)
    
    return dist_p
```

### Chuẩn p=1

```python
dist_p1=pd.DataFrame(distp(1))
```

```python
dist_p1.head()
```

**dist_p1 là ma trận vuông nxn với n là số feature của dữ liệu ban đầu, ma trận đối xứng qua đường chéo vì d(x,y)=d(y,x). Và khoảng cách trên đường chéo chính là 0 vì d(x,x)=0.**

### Chuẩn p=2

```python
dist_p2=pd.DataFrame(distp(2))
```

```python
dist_p2.head()
```

```python
from sklearn.preprocessing import KBinsDiscretizer
```

```python
discretizer=KBinsDiscretizer(n_bins=10,encode='ordinal', strategy='uniform' )
```

```python
df_equi_with=pd.DataFrame(discretizer.fit_transform(df2.values),columns=df2.columns)
```

Cảnh báo về feature 16 

Xét 2 bản ghi $ X =(𝑥_1 … 𝑥_𝑑) $ và  $ Y = (𝑦_1 … 𝑦_𝑑) $, sự tương đồng đơn giản nhất giữa 2 
bản ghi này được xác định như sau:
$$ Sim(x_i, y_i)= 
\sum_{i=1}^d S(x_i,y_i)
 $$ 

Với độ đo **inverse occurrence frequency** ta có công thức tính 
$$ S(x_i,y_i)=  
\begin{cases}
 \frac{1}{ p_k(x_i)^2}, x_i=y_i\\
  0 \ x_i \ne y_1
\end{cases}
$$

Nên ta có hàm sau:

```python
def inv_occur_frequency_measure(x,y):
    
    a,b=df_equi_with.shape
    index=df_equi_with.columns
    Sum=0;
    for i in range(0,b):
        if x[i]==y[i]:
            count=df_equi_with[index[i]].value_counts()
            pk=count[x[i]]/len(df_equi_with[index[i]]) # tính P_k(x_i)
            Sum=Sum+1/(pk**2)
    return Sum
            
```

Với tham số của hàm inv_occur_frequency là hai bản ghi X,Y 

```python
# ta tính bản ghi 0 và 1
inv_occur_frequency_measure(df_equi_with.iloc[0].values,df_equi_with.iloc[1].values)
```

Với độ đo **overlap/match** ta có công thức tính:
$$ S(x_i,y_i)=\begin{cases}
 1, x_i=y_i\\
  0 \ x_i \ne y_1
\end{cases} $$
Do đó ta có hàm

```python
def match_measure(x,y):
    count=0
    a,b=df_equi_with.shape
    for i in range(0,b):
        if x[i]==y[i]:
            count=count+1
    
    return count
```

```python
# độ tương đồng giữa bản ghi 0 và 1
match_measure(df_equi_with.iloc[0].values,df_equi_with.iloc[0].values)
```

### Tính láng giềng gần nhất cho điểm dữ liệu

Hàm tính láng giềng gần theo độ đo match measure 
* Đầu vào hàm là 1 điểm dữ liệu **point** và tập điểm dữ liệu **points** cần tính láng giềng và **k** điểm láng giềng


```python

def nearest_neighbor_match_measure(point, points,k):

    measure_match=[]
    for p in points:
        measure_match.append(match_measure(point,p)) # lưu độ đo vào 1 araay
    sorted_measure=np.argsort(measure_match) # lưu các vị trí sau khi xắp xếp tăng so với ban đầu


    k_nearest=sorted_measure[-k:] # lấy K điểm cuối do dữ liệu xếp tăng
    return points[k_nearest] # Kết quả trả về các bản ghi có láng giềng gần với point
```

Ta tính láng giềng bản ghi hàng thứ nhất trong bộ bản ghi từ hàng thứ 2 đến 10000 và lấy 3 bộ

```python

point=df_equi_with.iloc[0].values
points=df_equi_with.iloc[1:1000].values
k=3
nearest_neighbor_match_measure(point, points,k)
```

Tương tự ta có hàm tính láng giềng gần theo độ đo **inverse occurrence frequency measure**:
với **point** là điểm cần tìm kiếm láng giềng, **points** là bộ dữ liệu để tìm láng giềng, **k** là số bộ cần tìm láng giềng

```python

def nearest_neighbor_iof_measure(point, points,k):

    measure_iof=[]
    for p in points:
        measure_iof.append(inv_occur_frequency_measure(point,p))
    sorted_measure=np.argsort(measure_iof)

    
    k_nearest=sorted_measure[:k]
    return points[k_nearest]
```

```python
point=df_equi_with.iloc[0,:].values
points=df_equi_with.iloc[0:3,:].values
k=3
nearest_neighbor_iof_measure(point,points,k)
```
